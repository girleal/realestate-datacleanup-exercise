#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

```python
copiaCS = ds.copy()
filtroCS1 = copiaCS['level5'].isin(['Fuenlebrada', 'Leganés', 'Getafe', 'Alcorcón'])
cinturonSur = copiaCS[filtroCS1]

for x, y in cinturonSur.iterrows():
    ##min_price = copiaNormalizada['price'].min()
    ##max_price = copiaNormalizada['price'].max()
    ##copiaNormalizada['price_norm'] = (copiaNormalizada['price'] - min_price) / (max_price - min_price)
    #Cuando quieres valores entre 0 y 1, ideal para redes neuronales
    media = cinturonSur['price'].mean()
    desv = cinturonSur['price'].std()
    cinturonSur['price_z'] = (cinturonSur['price'] - media) / desv
    #Cuando hay outliers o vas a usar algoritmos que suponen distribución normal (ej. regresión)
print(cinturonSur[['price', 'price_z']].head())



poblaciones = ['Fuenlebrada', 'Leganés', 'Getafe', 'Alcorcón']
fig, axN = plt.subplots(2, 2, figsize=(12, 8))
axN = axN.flatten()

for i, poblacion in enumerate(poblaciones):
    datos = cinturonSur[cinturonSur['level5'] == poblacion]['price_z']
    axN[i].hist(datos, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
    axN[i].set_title(poblacion)
    axN[i].set_xlabel("Precio normalizado (Z-score)")
    axN[i].set_ylabel("Cantidad")
    axN[i].grid(True)

fig.suptitle("Distribución de precios normalizados por población", fontsize=14)
fig.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
## Aquí: traza la coordenadas de los estados

## PON TU CÓDIGO AQUÍ:

ds['lat'] = ds['latitude'].str.replace(',', '.', regex=False).astype(float)
ds['lon'] = ds['longitude'].str.replace(',', '.', regex=False).astype(float)

# Filtrar por poblaciones del Cinturón Sur
poblaciones = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']
cinturonSur = ds[ds['level5'].isin(poblaciones)].copy()

# Eliminar coordenadas inválidas (por ejemplo, 0 o NaN)
cinturonSur = cinturonSur[(cinturonSur['lat'] > 35) & (cinturonSur['lon'] < 0)]

# Crear gráfico limpio
colores = {
    'Fuenlabrada': 'red',
    'Leganés': 'blue',
    'Getafe': 'green',
    'Alcorcón': 'purple'
}

plt.figure(figsize=(10, 8))
for poblacion in poblaciones:
    data = cinturonSur[cinturonSur['level5'] == poblacion]
    plt.scatter(data['lon'], data['lat'], label=poblacion, color=colores[poblacion], alpha=0.6)

plt.title('Coordenadas de propiedades por población (Cinturón Sur de Madrid)')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```
